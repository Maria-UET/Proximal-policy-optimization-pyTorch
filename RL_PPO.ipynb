### General structure of the project:

env =gym.make(your_env) # like 'CartPole-v0'

def myRLmethod():

    for loop for number of episodes:
        env.reset() # reset after every episode? 

        for loops for number of steps or while loop that breaks when env is done (e.g. agent hits a deadend)
            env.render() # render after every step?
            action = env.action_space.getaction # get action  from action space
            observation, reward, done, info = env.step(action)# take action
            #print(observation,reward, done, info)# observation is the observ of next state of env after taking action
            if done:
                break        

myRLmethod()

```python
# run for 20 episodes each containing 100 steps, if the pole falls, start new episode
# take random actions

import gym
env = gym.make('CartPole-v0')
for i_episode in range(20): 
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()# random action
        observation, reward, done, info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```
